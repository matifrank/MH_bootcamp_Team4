```python
import os

folder_path = ' MH_bootcamp_Team4/Task10/'

dfs = {}

for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_path, dtype=str)
    elif file_name.endswith('.xlsx'):
        df = pd.read_excel(file_path, dtype=str)
    else:
        continue

    df_name = os.path.splitext(file_name)[0]
    dfs[df_name] = df
```

```python
dfs['ambito']['Fecha'] = pd.to_datetime(dfs['ambito']['Fecha'], errors='coerce')
dfs['blue']['Fecha'] = pd.to_datetime(dfs['blue']['Fecha'], errors='coerce')
dfs['divisa']['Fecha'] = pd.to_datetime(dfs['divisa']['Fecha'], errors='coerce')
dfs['ICL']['Fecha'] = pd.to_datetime(dfs['ICL']['Fecha'], errors='coerce')
dfs['REM']['Fecha'] = pd.to_datetime(dfs['REM']['Fecha'], errors='coerce')
dfs['UVA']['Fecha'] = pd.to_datetime(dfs['UVA']['Fecha'], errors='coerce')
dfs['BNA']['Fecha'] = pd.to_datetime(dfs['BNA']['Fecha'], errors='coerce')
```

```python
dfs['ambito']['Dólar Ambito'] = dfs['ambito']['Dólar Ambito'].str.replace(',', '.')
dfs['ICL']['ICL'] = dfs['ICL']['ICL'].str.replace(',', '.')
dfs['REM']['REM'] = dfs['REM']['REM'].str.replace(',', '.')
dfs['UVA']['UVA'] = dfs['UVA']['UVA'].str.replace(',', '.')
dfs['divisa']['Dólar Divisa BNA'] = dfs['divisa']['Dólar Divisa BNA'].str.replace(',', '.')
```

```python
dfs['blue'] = dfs['blue'].drop(columns=['Unnamed: 0'])
```

```python
dfs['blue'][' Dólar Blue '] = dfs['blue'][' Dólar Blue '].str.replace(',', '.')
```

```python

file_path = ' MH_bootcamp_Team4/Task10/indi/Indicadores.csv'
df = pd.read_csv(file_path, encoding='latin1')
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')
df.columns = [column.strip() for column in df.columns]
df = df.replace(' -   ', np.nan)
```

```python
df.columns
```

```python
import numpy as np

merged_df = df.merge( dfs['ambito'] , on='Fecha', how='left')
```

```python
merged_df['Dólar Ambito_x'].fillna(merged_df['Dólar Ambito_y'], inplace=True)
merged_df.drop(columns=['Dólar Ambito_y','Euro BNA','Dólar Ahorro','Diesel','Nafta'], inplace=True)
```

```python
merged_df
```

```python
merged_df2 = merged_df.merge( dfs['blue'] , on='Fecha', how='left')
#dict_keys(['ambito', 'blue', 'BNA', 'divisa', 'ICL', 'REM', 'UVA'])
merged_df2['Dólar Blue'].fillna(merged_df2[' Dólar Blue '], inplace=True)
merged_df2.drop(columns=[' Dólar Blue '], inplace=True)
merged_df2
```

```python
merged_df3 = merged_df2.merge( dfs['divisa'] , on='Fecha', how='left')
merged_df3['Dólar Divisa BNA_x'].fillna(merged_df3['Dólar Divisa BNA_y'], inplace=True)
merged_df3.drop(columns=['Dólar Divisa BNA_y' ], inplace=True)

merged_df3
```

```python
merged_df4 = merged_df3.merge( dfs['BNA'] , on='Fecha', how='left')
merged_df4['Dólar BNA_x'].fillna(merged_df4['Dólar BNA_y'], inplace=True)
merged_df4.drop(columns=['Dólar BNA_y'], inplace=True)

merged_df4
```

```python
merged_df5 = merged_df4.merge( dfs['ICL'] , on='Fecha', how='left')
merged_df5['ICL_x'].fillna(merged_df5['ICL_y'], inplace=True)
merged_df5.drop(columns=['ICL_y'], inplace=True)

merged_df5
```

```python
merged_df6 = merged_df5.merge( dfs['UVA'] , on='Fecha', how='left')
merged_df6['UVA_x'].fillna(merged_df6['UVA_y'], inplace=True)
merged_df6.drop(columns=['UVA_y'], inplace=True)

merged_df6
```

```python
#
dfs['REM'].set_index('Fecha', inplace=True)

dfs['REM_daily'] = dfs['REM'].resample('D').ffill()

dfs['REM_daily'].reset_index(inplace=True)
```

```python
merged_df7 = merged_df6.merge( dfs['REM_daily'] , on='Fecha', how='left')
merged_df7['REM_x'].fillna(merged_df7['REM_y'], inplace=True)
merged_df7.drop(columns=['REM_y'], inplace=True)

merged_df7
```

```python
merged_df7[['Dólar BNA_x', 'Dólar Divisa BNA_x','Dólar Ambito_x','Dólar Blue']] = merged_df7[['Dólar BNA_x', 'Dólar Divisa BNA_x','Dólar Ambito_x','Dólar Blue']].fillna(method='bfill')
```

```python

merged_df7.columns = merged_df7.columns.str.replace('x', '').str.replace('y', '').str.replace('_', '').str.lower()
merged_df7.rename(columns=lambda x: x.replace('dólar', 'dolar'), inplace=True)
```

```python
merged_df7.columns = merged_df7.columns.str.strip()
merged_df7.set_index('fecha', inplace=True)
merged_df7.to_csv('indicadores_economicos_AR-transform.csv')
```
